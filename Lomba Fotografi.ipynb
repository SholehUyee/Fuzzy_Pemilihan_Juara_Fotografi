```python
def FungsiKeanggotaan(_range, _lo, _mi, _hi, _nilai):
    lo = fuzz.interp_membership(_range, _lo, _nilai)
    mi = fuzz.interp_membership(_range, _mi, _nilai)
    hi = fuzz.interp_membership(_range, _hi, _nilai)
    
    return lo, mi, hi
```
